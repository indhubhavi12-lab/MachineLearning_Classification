```python
import pickle
import numpy as np

# Load model
loaded_model = pickle.load(open("finalized_model_SVM.sav", 'rb'))

# Load scaler
loaded_scaler = pickle.load(open("scaler_SVM.sav", 'rb'))
```

```python
preinput_data = np.array([[45, 80, 1.020, 1, 0, 1, 1, 0, 0,
                           150, 40, 1.2, 135, 4.5, 15,
                           40, 8000, 4.5, 1, 1, 0, 1, 0, 0,
                           0, 1, 0]])
```

```python
# Scale input
preinput = loaded_scaler.transform(preinput_data)

# Predict
result = loaded_model.predict(preinput)

print("Prediction:", result)
```
