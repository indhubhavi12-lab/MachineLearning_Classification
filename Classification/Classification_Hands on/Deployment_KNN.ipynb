```python
import pickle
import numpy as np

# Load model
loaded_model = pickle.load(open("finalized_model_KNN.sav", "rb"))

# Load scaler
loaded_scaler = pickle.load(open("scaler_KNN.sav", "rb"))

# New input data
preinput_data = np.array([[22, 20000, 1]])

# Scale input
preinput_scaled = loaded_scaler.transform(preinput_data)

# Predict
result = loaded_model.predict(preinput_scaled)

# Output result
if result[0] == 1:
    print("Customer Will Purchase")
else:
    print("Customer Will NOT Purchase")
```
